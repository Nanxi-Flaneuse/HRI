```python
import pandas as pd
import numpy as np
from scipy.stats import shapiro
from scipy import stats
```

```python
time1 = np.array([37, 40, 40, 21.88])
time2 = np.array([49, 48, 40, 57, 20])

print(np.mean(time1))
print(np.mean(time2))
```

```python
print(shapiro(time1))
print(shapiro(time2))
```

```python
# group1_data = np.array([17, 19, 15, 13, 18, 12, 15, 16, 13])
# group2_data = np.array([16, 15, 17, 13, 14, 12, 15, 18, 11])

# Perform Mood's Median Test
# The function takes the samples as separate arguments
from scipy.stats import median_test
p_value = median_test(time1, time2)


# Print the results
# print(f"Chi-squared statistic: {statistic}")
print(f"P-value: {p_value}")
```
